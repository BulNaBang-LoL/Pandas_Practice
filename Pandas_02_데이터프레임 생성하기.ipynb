```python
import pandas as pd
```

```python
friend_dict_list = [
                    {'name': 'John', 'age': 25, 'job': 'student'},
                    {'name': 'Nate', 'age': 30, 'job': 'teacher'}
]
```

```python
df = pd.DataFrame(friend_dict_list)
```

```python
df.head()
```

dictionary의 경우 column의 순서가 랜덤이다.

지금은 우연히 맞았지만, 

만약 순서 변경이 필요한 경우에는 직접 순서를 명시해주는 방법으로 해결할 수 있다.

(예) df = df[['name', 'age', 'job']]

아니면 ordered dictionary를 사용하는 방법도 있다.

```python
from collections import OrderedDict
```

```python
friend_ordered_dict = OrderedDict(
    [
     ('name', ['John', 'Nate']),
     ('age', ['25', '30']),
     ('job', ['student', 'teacher'])
    ]
)
```

```python
df = pd.DataFrame.from_dict(friend_ordered_dict)
```

```python
df.head()
```

```python
friend_list = [
               ['John', 25, 'student'],
               ['Nate', 30, 'teacher']
]
```

```python
column_name = ['name', 'age', 'job']
```

```python
df = pd.DataFrame.from_records(friend_list, columns = column_name)
```
